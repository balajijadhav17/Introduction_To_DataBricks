## **Create DF using list and tuple**

```python
spark.createDataFrame([("balaji",1)]).show()
```

```python
data=[(1,"Maheer"),(2,"jadhav"),(3,"balu")]
df=spark.createDataFrame(data=data)
df.show()
```

```python
from pyspark.sql.types import *
data=[(1,"Maheer"),(2,"jadhav"),(3,"balu")]


df=spark.createDataFrame(data=data,schema=["id","name"])
df.show()
df.printSchema()
```

```python
from pyspark.sql.types import *

data=[(1,"Maheer"),(2,"jadhav"),(3,"balu")]

schema=StructType([StructField(name="id",dataType=IntegerType()),
            StructField(name="name",dataType=StringType())])

df=spark.createDataFrame(data=data,schema=schema)
df.show()
df.printSchema()
```

## **DF using Dict**

```python
data=[{"id":1,"name":"Maheer"},{"id":2,"name":"jadhav"},{"id":3,"name":"balu"}]
df1=spark.createDataFrame(data=data)
df1.show()
df1.printSchema()
```

## **DF using Tuple**

```python
data=[(1,"bj"),(2,"rj")]
schema=["id","name"]
df3=spark.createDataFrame(data,schema)
df3.show()
df3.printSchema()
```

```sql
with join_query
as
(
select s.total_amount,p.product_name from sales_enr s 
left join products_enr p on s.product_id = p.product_id)

select * from join_query;

-- apply group by

with join_query
as
(
select sum(s.total_amount) as total_Amount ,p.product_name from sales_enr s 
left join products_enr p on s.product_id = p.product_id
group by p.product_name)

select * from join_query;

-- apply dense rank

with join_query
as
(
select sum(s.total_amount) as total_Amount ,p.product_name from sales_enr s 
left join products_enr p on s.product_id = p.product_id
group by p.product_name)

select * , dense_rank() over(order by total_Amount desc) as ranking from join_query;


-- Top 5 total amount product name

with join_query
as
(
select sum(s.total_amount) as total_Amount ,p.product_name from sales_enr s 
left join products_enr p on s.product_id = p.product_id
group by p.product_name)

select * , dense_rank() over(order by total_Amount desc) as ranking from join_query limit 5;


-- Top 5 using "add parameter function" and nested CTE

with join_query
as
(
select sum(s.total_amount) as total_Amount ,p.product_name from sales_enr s 
left join products_enr p on s.product_id = p.product_id
group by p.product_name),
rank_query
as
(select * , dense_rank() over(order by total_Amount desc) as ranking from join_query)

select * from rank_query where ranking <= :top;





```

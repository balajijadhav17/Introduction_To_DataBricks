# **UDF** (USER DEFINED FUNCTIONS)

```python
def greet(p_input):
    return "@"+str(p_input)
```

```python
udf_greet=udf(greet)
```

```python
from pyspark.sql.functions import col
df=df.withColumn("greet",udf_greet(col("domain")))
display(df)
```

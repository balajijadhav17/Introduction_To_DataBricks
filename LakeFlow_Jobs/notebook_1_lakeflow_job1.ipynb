# Read **Data**

```python
df=spark.read.format("csv") \
    .option("header","true")\
    .option("inferSchema","true")\
    .load("/Volumes/my_first_catalog/bronze/bronze_volume2/stores/")
```

# Write **Data**

```python
df.write.format("delta")\
    .mode("overwrite")\
    .save("/Volumes/my_first_catalog/bronze/jobvolume/raw/stores/")
```
